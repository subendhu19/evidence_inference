```python
import re
import pickle
import unicodedata

annotation_in_paragraphs = 0
all_instances = 0
regex = re.compile('[^a-zA-Z]')

par_split_annotations = {}

for i in range(len(annotations)):
    if not annotations['Valid Label'][i] or (not annotations['Valid Reasoning'][i]):
        continue
    pmc_id = annotations['PMCID'][i]
    prompt_id = annotations['PromptID'][i]
    
    if pmc_id not in par_split_annotations:
        par_split_annotations[pmc_id] = {}
        
    if not prompt_id in par_split_annotations[pmc_id]:
        par_split_annotations[pmc_id][prompt_id] = 'set'
    
    annotation = regex.sub('', str(annotations['Annotations'][i]).lower())
    
    article = etree.parse('../data/xml_files/PMC' + str(pmc_id) + '.nxml')
    
    paragraphs = []
    for element in article.getroot():
        if element.tag == 'front':
            for sec_element in element:
                if sec_element.tag == 'article-meta':
                    for abs_element in sec_element:
                        if abs_element.tag == 'abstract':
                            for sec_element in abs_element:
                                if sec_element.tag == 'p':
                                    par_text = etree.tostring(sec_element, encoding='utf8', method='text').decode("utf-8") 
                                    paragraphs.append(par_text)
                                else:
                                    for par_element in sec_element:
                                        par_text = etree.tostring(par_element, encoding='utf8', method='text').decode("utf-8") 
                                        paragraphs.append(par_text)
        if element.tag == 'body':
            for sec_element in element:
                for par_element in sec_element:
                    par_text = etree.tostring(par_element, encoding='utf8', method='text').decode("utf-8") 
                    paragraphs.append(par_text)
    
    present = False
    evidence_idx = -1
    for p in paragraphs:
        evidence_idx += 1
        norm_p = regex.sub('', str(p).lower())
        if annotation in norm_p:
            present = True
            break
    
    if not present:
        evidence_idx = -1
        
    if par_split_annotations[pmc_id][prompt_id] == 'set':
        par_split_annotations[pmc_id][prompt_id] = [paragraphs, annotations['Label'][i], [evidence_idx]]
    else:
        par_split_annotations[pmc_id][prompt_id][2].append(evidence_idx)

all_questions = {}
for pmc_id in par_split_annotations:
    all_questions[pmc_id] = []
    for prompt_id in par_split_annotations[pmc_id]:
#         print(list(set(par_split_annotations[pmc_id][prompt_id][2])))
        evidence_indices = [item for item in list(set(par_split_annotations[pmc_id][prompt_id][2])) if item != -1]
#         print(evidence_indices)
        if len(evidence_indices) > 0:
            all_questions[pmc_id].append(par_split_annotations[pmc_id][prompt_id][:-1] + [evidence_indices, prompt_id])
            annotation_in_paragraphs += 1
        all_instances += 1
        
pickle.dump(all_questions, open('../data/data/p_annotations.p', 'wb'))
print(annotation_in_paragraphs)
print(all_instances)
```

```python
import numpy as np

def get_overlap_ratio(l1, l2):
    overlap = 0
    for word in l2:
        if word in l1:
            overlap += 1
    return overlap*1.0/len(l2)

overlap_ratios_1 = []
overlap_ratios_2 = []

for i in range(len(annotations['Annotations'])):
    annotation = unicode(str(annotations['Annotations'][i]).lower(), 'ascii', 'ignore').replace('\t', ' ')
    prompt = unicode(prompt_dict[annotations['PromptID'][i]].lower(), 'ascii', 'ignore')
    
    ann_words = [item for item in annotation.split(' ') if item != u'']
    p_words = prompt.split(' ')
    
    overlap_ratio_1 = get_overlap_ratio(ann_words, p_words)
    overlap_ratio_2 = get_overlap_ratio(p_words, ann_words)
    
    overlap_ratios_1.append(overlap_ratio_1)
    overlap_ratios_2.append(overlap_ratio_2)
    
print 'Average overlap ratio (prompt):', np.mean(overlap_ratios_1)
print 'Average overlap ratio (annotation):', np.mean(overlap_ratios_2)
    
```

```python
import matplotlib.pyplot as plt

bins = range(11)
bin_counts = {}

for b in bins:
    bin_counts[b] = 0
    
for r in overlap_ratios_2:
    bin_counts[int(r*10)] += 1
    
plt.plot(np.array(bin_counts.keys())*0.1, bin_counts.values())
plt.ylabel('Number of samples')
plt.xlabel('Ratio')
plt.show()
```

```python
import numpy as np

def get_one_hot(lid, vocab_size):
    ret = np.zeros(vocab_size)
    for i in lid:
        ret[i] += 1
    return ret
    
data = [(0, 'A', 'I am Sub', 'PTYY'), (1, 'B', 'I am Em', 'ADSC'), (2, 'A', 'Sub am Em', 'SERT')]

bats = [item[1] for item in data]
codes = [item[3] for item in data]
words = [item[2].split(' ') for item in data]

vocab_bats = set(bats)
vocab_codes = set(codes)
vocab_words = set([item for sublist in words for item in sublist])

def get_feature_vector(sample):
    bid = list(vocab_bats).index(sample[1])
    cid = list(vocab_codes).index(sample[3])
    tids = [list(vocab_words).index(item) for item in sample[2].split(' ')]
    bl= get_one_hot([bid], len(vocab_bats))
    cl= get_one_hot([cid], len(vocab_codes))
    tl= get_one_hot(tids, len(vocab_words))
    return list(bl) + list(tl) + list(cl)

print vocab_bats, vocab_words, vocab_codes
print np.array(data)
print np.array([get_feature_vector(item) for item in data])

```
